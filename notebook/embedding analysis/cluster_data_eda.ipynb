```python
import pandas as pd
import json
import pickle
from huggingface_hub import hf_hub_download
from sklearn.model_selection import train_test_split
```

```python
# Load the data from the uploaded file
# file_path = '../data/enhanced_PR_dataset_with_embeddings.json'
file_path = '../data/enhanced_HER2_dataset_with_embeddings_1_without_scaling.json'

# Read the JSON file into a dataframe
with open(file_path, 'r') as f:
    data = json.load(f)

# Convert the JSON data into a DataFrame
df = pd.DataFrame(data)

# Filter the rows where 'predicts_correct' is False
filtered_df = df[df['predicts_correct'] == False]
```

```python
filtered_df.rename(columns={'sample_id': 'slide_id'}, inplace=True)
```

```python
slide_feature_path = hf_hub_download(
    "MahmoodLab/TITAN", 
    filename="TCGA_TITAN_features.pkl",
)
with open(slide_feature_path, 'rb') as file:
  data = pickle.load(file)
embeddings_df = pd.DataFrame({'slide_id': data['filenames'], 'embeddings': list(data['embeddings'][:])})
```

```python
cluster_data = pd.merge(embeddings_df, filtered_df, on='slide_id')
```

```python
cluster_data.to_csv('../data/ERBB2_cluster_data_without_scaling_stratify.csv')
```
